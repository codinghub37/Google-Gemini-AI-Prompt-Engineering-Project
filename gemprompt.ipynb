```python
import google.generativeai as genai

genai.configure(api_key=" ")

model = genai.GenerativeModel("models/gemini-2.5-flash")

response = model.generate_content("What is Artificial Intelligence?")

print(response.text)
```

```python
response = model.generate_content("hello how are You")

print(response.text)
```

```python
response = model.generate_content("Act as an expert travel agent. Create a detailed 3-day itinerary for a trip to Paris focusing on art and history. Provide the output in bullet points for each day.")

print(response.text)
```
